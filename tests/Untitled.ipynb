```python
import os
from datetime import datetime,timedelta
import random
import math
import numpy as np
import pandas as pd
import cnmaps
from cnmaps import get_adm_maps, draw_maps
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
```

```python
stations_df=pd.read_csv('../datas/sts.csv')
```

```python
def query_location(point, geometries, names):
    """查询点所在的行政区划"""
    for i, geom in enumerate(geometries):
        if isinstance(geom, list):  # 处理多部分多边形
            for part in geom:
                if part.contains(point):
                    return names[i]
        else:  # 单部分多边形
            if geom.contains(point):
                return names[i]
    return None
```

```python
# 获取省级边界
provinces = get_adm_maps(level='省', only_polygon=True)
province_names = [p['省/直辖市'] for p in get_adm_maps(level='省')]
```

```python
# 获取市级边界
cities = get_adm_maps(level='市', only_polygon=True)
city_names = [c['市'] for c in get_adm_maps(level='市')]
```

```python
# 获取县级边界
counties = get_adm_maps(level='区县', only_polygon=True)
county_names = [c['区/县'] for c in get_adm_maps(level='区县')]
```

```python
# 为每个站点添加位置信息
stations_df['Province'] = None
stations_df['City'] = None
stations_df['County'] = None
```

```python
%%time
# 遍历每个站点进行查询
for index, row in stations_df.iterrows():
    point = Point(row['Lon'], row['Lat'])
    
    # 查询省份
    province = query_location(point, provinces, province_names)
    stations_df.loc[index, 'Province'] = province
    
    # 查询城市
    city = query_location(point, cities, city_names)
    stations_df.loc[index, 'City'] = city
    
    # 查询区县
    county = query_location(point, counties, county_names)
    stations_df.loc[index, 'County'] = county
    
    print(f"站点 {row['Station_Id_C']} 位置查询完成")
```

```python
def read_province_data(file_path):
    """读取省份数据并构建名称到编码的映射"""
    df = pd.read_csv(file_path, skipinitialspace=True)
    # 处理第一列可能是索引的情况
    if 'code' not in df.columns and 'name' not in df.columns:
        df = pd.read_csv(file_path, header=None, skipinitialspace=True)
        df.columns = ['index', 'code', 'name'] if len(df.columns) == 3 else ['code', 'name']
    
    # 构建映射字典，处理名称中的空格
    province_map = {name.strip(): code for code, name in zip(df['code'], df['name'])}
    return province_map

def read_city_data(file_path):
    """读取城市数据并构建名称到编码的映射"""
    df = pd.read_csv(file_path, skipinitialspace=True)
    # 处理第一列可能是索引的情况
    if 'code' not in df.columns and 'name' not in df.columns:
        df = pd.read_csv(file_path, header=None, skipinitialspace=True)
        df.columns = ['index', 'code', 'name', 'provinceCode'] if len(df.columns) == 4 else ['code', 'name', 'provinceCode']
    
    # 构建映射字典，处理名称中的空格
    city_map = {}
    for _, row in df.iterrows():
        city_name = row['name'].strip()
        province_code = row['provinceCode']
        city_code = row['code']
        # 使用 (省份编码, 城市名称) 作为键，确保唯一性
        city_map[(province_code, city_name)] = city_code
    return city_map

def read_county_data(file_path):
    """读取区县数据并构建名称到编码的映射"""
    df = pd.read_csv(file_path, skipinitialspace=True)
    # 处理第一列可能是索引的情况
    if 'code' not in df.columns and 'name' not in df.columns:
        df = pd.read_csv(file_path, header=None, skipinitialspace=True)
        df.columns = ['index', 'code', 'name', 'cityCode', 'provinceCode'] if len(df.columns) == 5 else ['code', 'name', 'cityCode', 'provinceCode']
    
    # 构建映射字典，处理名称中的空格
    county_map = {}
    for _, row in df.iterrows():
        county_name = row['name'].strip()
        city_code = row['cityCode']
        province_code = row['provinceCode']
        # 使用 (省份编码, 城市编码, 区县名称) 作为键，确保唯一性
        county_map[(province_code, city_code, county_name)] = row['code']
    return county_map
```

```python
province_map = read_province_data('../datas/provinces.csv')
city_map = read_city_data('../datas/cities.csv')
county_map = read_county_data('../datas/areas.csv')
```

```python
stations_df['ProvinceCode'] = None
stations_df['CityCode'] = None
stations_df['CntyCode'] = None
```

```python
for index, row in stations_df.iterrows():
    province_name = row['Province'].strip()
    city_name = row['City'].strip()
    county_name = row['Cnty'].strip()
    
    # 获取省份编码
    province_code = province_map.get(province_name)
    stations_df.loc[index, 'ProvinceCode'] = province_code
    
    if province_code:
        # 获取城市编码
        city_code = city_map.get((province_code, city_name))
        stations_df.loc[index, 'CityCode'] =str(city_code)
        
        if city_code:
            # 获取区县编码
            county_code = county_map.get((province_code, city_code, county_name))
            stations_df.loc[index, 'CntyCode'] = str(county_code)
```

```python
stations_df=stations_df.dropna(subset='CntyCode').reset_index(drop=True)
```

```python
stations_df.to_csv('../datas/Stations.csv',index=False)
```
